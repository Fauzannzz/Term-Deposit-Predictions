*   Customers with a previous contact (`hari_sejak_kontak_sebelumnya` != 999) have a significantly higher subscription rate (around 64%) compared to those with no previous contact (around 9%).

*   Subscribers tend to have fewer contacts in the current campaign (`jumlah_kontak_kampanye_ini`) compared to non-subscribers.
*   Customers with 'mahasiswa', 'pensiunan', and 'unemployed' occupations, and those with 'TIDAK SEKOLAH' or 'Pendidikan Tinggi' education levels show relatively higher subscription rates.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, f1_score

from xgboost import XGBClassifier
```

```python
# Load data
df = pd.read_csv("training_dataset.csv")  # Ganti dengan path Anda jika belum

# Drop identifier
df.drop(columns=['customer_number'], inplace=True)

# Tangani missing value jika ada
df.replace('unknown', np.nan, inplace=True)
df.dropna(inplace=True)

# Konversi target ke int
df['berlangganan_deposito'] = df['berlangganan_deposito'].astype(int)

# Pisahkan fitur dan target
X = df.drop('berlangganan_deposito', axis=1)
y = df['berlangganan_deposito']

# Label encoding untuk fitur kategorikal
categorical_cols = X.select_dtypes(include='object').columns
X_encoded = X.copy()

for col in categorical_cols:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X[col])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=5,
    scale_pos_weight=(y_train == 0).sum() / (y_train == 1).sum(),  # untuk imbalance
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
print("F1 Score:", f1_score(y_test, y_pred))
print("Accuracy:", (y_pred == y_test).mean())
print("recall:", (y_pred[y_test == 1] == 1).mean())
print("precision:", (y_pred[y_test == 1] == 1).sum() / (y_pred == 1).sum())
```
